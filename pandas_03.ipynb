```python
#Concatenation


import pandas as pd

df1 = pd.read_csv('test1.csv')

df2 = pd.read_csv('test3.csv')

df3 = pd.read_csv('update.csv')
```

```python
row_conc=pd.concat([df1,df2,df3])
row_conc
```

```python
#Access perticuler row,
print(row_conc.iloc[4,])
```

```python
# create a new row using series

new_row=pd.Series(['Blore','Mandya','Vizak','Bopal','UK'])
new_row
```

```python
# adding new rows
print(pd.concat([df1,new_row]))
```

```python
new_row_df = pd.DataFrame([['Blore','Mandya','Vizak','Bopal','UK']], columns=['A', 'B', 'C', 'D', 'E'])
new_row_df
```

```python
print(pd.concat([df1, new_row_df]))
```

```python
#append

print(df1.append(df2))
```

```python

# concat and append are the same
row_concat_i = pd.concat([df1, df2, df3], ignore_index=True)
row_concat_i
```

```python
col_concat = pd.concat([df1, df2, df3], axis=1)
col_concat
```

```python
#add new single column: with list
col_concat['new_col'] = ['Bhavya', 'Udhu', 'Gopa', 'Ankit']
col_concat
```

```python
#add a series

col_concat['New_col_series']=pd.Series(['Raghav','Subhas','Dasa','Kala'])
col_concat
```

```python
print(pd.concat([df1, df2, df3], axis=1, ignore_index=1))
```

```python
#merging multiple data files


person = pd.read_csv('Person.csv')
site = pd.read_csv('Site.csv')
survey = pd.read_csv('Survey.csv')
visited = pd.read_csv('Visited.csv')
person
```

```python
site
```

```python
survey
```

```python
visited
```

```python
vis_subset = visited.loc[[0,4,6], ]
vis_subset
```

```python
site
```

```python
#one to one merge
o2o_merge = site.merge(vis_subset, left_on='Unnamed: 0', right_on='ident')
o2o_merge
```

```python
#many to one merge

m2o_merge = site.merge(visited, left_on='Unnamed: 0', right_on='ident')
m2o_merge
```

```python
#many to many merge
m2m_merge = person.merge(survey, left_on='Unnamed: 0', right_on='taken')
m2m_merge
```

```python
vs = visited.merge(survey, left_on='ident', right_on='taken')
vs
```

```python

ps_vs = m2m_merge.merge(vs,

                 left_on=['quant', 'reading'],

                 right_on=[ 'quant', 'reading'])

ps_vs
```
